6. Change column name from 'Confusion' to 'Dizzyness' and 'Smoking' to 'Smoker?'


```python
alzheimers_df.rename(columns={'Confusion': 'Dizzyness'}, inplace=True)
alzheimers_df.rename(columns={'Smoking': 'Smoker?'}, inplace=True)
alzheimers_df.tail(3)
```

7. Create a new column 'Gender Description' that displays "Male" if the value of Gender is 0, else "Female"

```python
alzheimers_df.insert(alzheimers_df.columns.get_loc('Gender') + 1, 'Gender Description', alzheimers_df['Gender'].apply(lambda x: 'Male' if x == 0 else 'Female'))
alzheimers_df.tail(3)
```

8.  Create a new dataframe 'Asian' that gathers data with Alzheimer's (In the Ethnicity column, Asian is coded as '2')

```python
asian_df = alzheimers_df[(alzheimers_df['Ethnicity'] == 2) & (alzheimers_df['Diagnosis'] == 1)]
asian_df.head(8)
```

9. Create a new dataframe 'Female' that gathers Female data without Alzheimer's

```python
female_df = alzheimers_df[(alzheimers_df['Gender'] == 1) & (alzheimers_df['Diagnosis'] == 0)]
female_df.head(8)
```

10. Create a new dataframe 'African-Americans' that gathers data of all African Americans in the dataset (In the Ethnicity column, African American is coded as '1')

```python
african_americans_df = alzheimers_df[alzheimers_df['Ethnicity'] == 1]
african_americans_df.head(5)
```

11. Create a new dataframe 'Uneducated' that gathers data of all patients that never finished high school (in the EducationalLevel column, it is coded as '0')


```python
uneducated_df = alzheimers_df[alzheimers_df['EducationLevel'] == 0]
uneducated_df.head(5)
```

12. Use numpy to get the average age and alcohol consumption


```python
average_age = np.mean(alzheimers_df['Age'])
average_alcohol_consumption = np.mean(alzheimers_df['AlcoholConsumption'])
print(f'Average Age: {average_age}')
print(f'Average Alcohol Consumption: {average_alcohol_consumption}')
```

13. Use numpy to get the median age and alcohol consumption


```python
median_age = np.median(alzheimers_df['Age'])
median_alcohol_consumption = np.median(alzheimers_df['AlcoholConsumption'])
print(f'Median Age: {median_age}')
print(f'Median Alcohol Consumption: {median_alcohol_consumption}')
```

14. Use numpy to get the middle values of age and alcohol consumption


```python
middle_age = np.median(alzheimers_df['Age'])
middle_alcohol_consumption = np.median(alzheimers_df['AlcoholConsumption'])
print(f'Middle Age: {middle_age}')
print(f'Middle Alcohol Consumption: {middle_alcohol_consumption}')
```

15. Use numpy to get the standard deviation of BMI


```python
std_bmi = np.std(alzheimers_df['BMI'])
print(f'Standard Deviation of BMI: {std_bmi}')
```

## Part 2. Descriptive Statistics and Data Visualizations of Possible Correlations between Alzheimer Diagnosis and other Factors

```python
# Summary statistics for each numerical variable
summary_stats = alzheimers_df.describe()
print(summary_stats)

# Frequency distribution for categorical variables
gender_distribution = alzheimers_df['Gender'].value_counts()
ethnicity_distribution = alzheimers_df['Ethnicity'].value_counts()
education_level_distribution = alzheimers_df['EducationLevel'].value_counts()

print("\nGender Distribution:")
print(gender_distribution)

print("\nEthnicity Distribution:")
print(ethnicity_distribution)

print("\nEducation Level Distribution:")
print(education_level_distribution)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Diagnosis' to a categorical variable for better plotting
alzheimers_df['Diagnosis'] = alzheimers_df['Diagnosis'].astype('category')

# Plot 1: Ethnicity vs. Diagnosis
plt.figure(figsize=(10, 6))
sns.countplot(x='Ethnicity', hue='Diagnosis', data=alzheimers_df)
plt.title('Ethnicity vs. Diagnosis')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.legend(title='Diagnosis')
plt.grid(True)
plt.show()

# Plot 2: BMI vs. Diagnosis
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diagnosis', y='BMI', data=alzheimers_df)
plt.title('BMI vs. Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('BMI')
plt.grid(True)
plt.show()

# Plot 3: Physical Activity vs. Diagnosis
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diagnosis', y='PhysicalActivity', data=alzheimers_df)
plt.title('Physical Activity vs. Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Physical Activity')
plt.grid(True)
plt.show()

# Plot 4: Education Level vs. Diagnosis
plt.figure(figsize=(10, 6))
sns.countplot(x='EducationLevel', hue='Diagnosis', data=alzheimers_df)
plt.title('Education Level vs. Diagnosis')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.legend(title='Diagnosis')
plt.grid(True)
plt.show()

# Plot 5: Alcohol Consumption vs. Diagnosis
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diagnosis', y='AlcoholConsumption', data=alzheimers_df)
plt.title('Alcohol Consumption vs. Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Alcohol Consumption')
plt.grid(True)
plt.show()
```
