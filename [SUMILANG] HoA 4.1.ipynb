For my second Tower of Hanoi implementation, I wanted to use a stack data structure because the puzzle is 3 stacks by nature that pops and appends elements from each other. 

I still used recursion because since there are 3 stacks by nature, it is guaranteed to move all the disks to the last stack in descending order.

```python
import time  # Import the time module to add delays

# Function to print the current state of the stacks
def print_stacks(A, B, C):
    print("A:", A)  # Print stack A
    print("B:", B)  # Print stack B
    print("C:", C)  # Print stack C
    print("\n")  # Print a newline for better readability

# Recursive function to solve the Tower of Hanoi problem
def TowerOfHanoi(n, source, destination, auxiliary, stacks):
    if n == 1:  # Base case: if there's only one disk to move
        disk = stacks[source].pop()  # Remove the top disk from the source stack
        stacks[destination].append(disk)  # Place the disk on the destination stack
        print(f"Move disk 1 from {source} to {destination}")  # Print the move
        print_stacks(stacks['A'], stacks['B'], stacks['C'])  # Print the current state of the stacks
        time.sleep(4)  # Delay for visualization (4 seconds)
        return  # Return from the function

    # Move n-1 disks from source to auxiliary, so they are out of the way
    TowerOfHanoi(n-1, source, auxiliary, destination, stacks)
    disk = stacks[source].pop()  # Remove the nth disk from the source stack
    stacks[destination].append(disk)  # Place the nth disk on the destination stack
    print(f"Move disk {n} from {source} to {destination}")  # Print the move
    print_stacks(stacks['A'], stacks['B'], stacks['C'])  # Print the current state of the stacks
    time.sleep(4)  # Delay for visualization (4 seconds)
    
    # Move the n-1 disks from auxiliary to destination
    TowerOfHanoi(n-1, auxiliary, destination, source, stacks)

# Driver code
n = 3  # Number of disks
stacks = {
    'A': list(range(n, 0, -1)),  # Initialize stack A with disks in descending order
    'B': [],  # Initialize stack B as empty
    'C': []  # Initialize stack C as empty
}

print_stacks(stacks['A'], stacks['B'], stacks['C'])  # Print the initial state of the stacks
TowerOfHanoi(n, 'A', 'B', 'C', stacks)  # Call the TowerOfHanoi function to solve the problem
```
